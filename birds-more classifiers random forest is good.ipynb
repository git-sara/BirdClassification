```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,confusion_matrix,classification_report, precision_recall_curve
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.model_selection import train_test_split

bird=pd.read_excel('C:\\Users\\Asus\\Desktop\\data sets\\bird2.xlsx')
#bird.head()
#sns.relplot(x="WingSpanRange", y="ConservationStatus", data=bird, height=5, aspect=.5)
#bird_train['ConservationStatus'].value_counts().plot.barh()#So imbalanced
bird=bird.drop(['MinLength','MaxLength','MinBodyMass','MaxBodyMass','MinWingspan','MaxWingspan'],axis=1)
bird.head()
X=bird.drop(['ConservationStatus'],axis=1)
y=bird['ConservationStatus']
oversample = SMOTE()
transformed_X, transformed_y= oversample.fit_resample(X, y)
transformed_df = pd.concat([transformed_X,transformed_y],axis=1, join='outer')
transformed_df.to_csv("C:\\Users\\Asus\\Desktop\\data sets\\bird3.csv")
transformed_df.head()

train_df,test_df=train_test_split(transformed_df, test_size=0.2, stratify=transformed_df['ConservationStatus'])
X_train=train_df.drop(['ConservationStatus'],axis=1)
y_train=train_df['ConservationStatus']
X_test=test_df.drop(['ConservationStatus'],axis=1)
y_test=test_df['ConservationStatus']


C = 10
# Create different classifiers.
classifiers = {
    'Linear SVC': SVC(kernel='linear', C=C, probability=True,random_state=0),
    'KNN classifier': KNeighborsClassifier(C),
    'SVC': SVC(),
    'RFST': RandomForestClassifier(n_estimators=100),
    'ADA': AdaBoostClassifier(n_estimators=100)
}

n_classifiers = len(classifiers)

for index, (name, classifier) in enumerate(classifiers.items()):
    classifier.fit(X_train, np.ravel(y_train))

    y_pred = classifier.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print("Accuracy (train) for %s: %0.1f%% " % (name, accuracy * 100))
    print(classification_report(y_test,y_pred))


    














```
